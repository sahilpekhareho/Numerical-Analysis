```python
from math import sqrt
def f(x,nmax=1000):
    for i in range(nmax):
        x = sqrt(x)
       # print("y=", x)
    for i in range(nmax):
        x = x**2
       # print("z=",x)
    return x
for xin in (5., 0.5):
    xout = f(xin)
    print(xin, xout)
```

Square rooting  a number greater than 1 several times brings it very close to 1 such that the number is essentially rounded off to 1 as it is 1 up to several decimal places.Squaring one,even doing so a hundred times we start getting 1 as $(1^2)^\infty=1$.

Similarly, for a number less than one, square rooting it brings it very close to but still less than that as python preserves the first 16 digits, and squaring that number a hundred times reduces it to zero.

```python
from math import sqrt
def bad_root_qdeqn(a,b,c):
    x_pl=(-b+sqrt((b**2)-4*a*c))/2*a
    x_mi=(-b-sqrt((b**2)-4*a*c))/2*a
    list=[x_pl,x_mi]
    return list
def good_root_qdeqn(a,b,c):
    x_mi=(-b-sqrt(b**2-4*a*c))/2*a
    x_pl=c/(x_mi*a)
    list=[x_pl,x_mi]
    return list
print(bad_root_qdeqn(1,10**8,1),good_root_qdeqn(1,10**8,1))
```

The answer for good root makes sense as the answer is just the reciprocal which should $-1\times10^-8$  as $\frac{c}{a}=1$, bad root answer is problematic as the radical and $b$ are very close $x_+$.

```python
import math as m
def sqwave(x):
    if x>-m.pi and x<=0:
        return -1/2
    elif x>0 and x<m.pi:
        return 1/2
def fourierseries(x,n):
    y=0
    for i in range(0,n+1):
        if i%2!=0:
            y+=m.sin(i*x)/i
    return (2/m.pi)*y
x=[-m.pi]
while x[len(x)-1] !=m.pi:
    x.append(x[len(x)-1]+0.05*m.pi)
```

```python
y1=[]
y1 = [sqwave(xs) for xs in x]
y2 = [fourierseries(xs,1)for  xs in x]
y3=[fourierseries(xs,3)for  xs in x]
y4=[fourierseries(xs,5)for  xs in x]
y5=[fourierseries(xs,7)for  xs in x]
y6=[fourierseries(xs,5000)for  xs in x]
```
